```python
import pandas as pd
```

```python
df_analysis = pd.read_csv('Data/df_analysis', index_col=[0])
df_analysis = df_analysis[['id', 'Gender', 'subtype', 'adhdtype', 'brain_oscillation', 'electrode', 'fft_abs_power']]
df_analysis
```

```python
df_Delta  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Delta']
df_Theta  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Theta']
df_Alpha  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Alpha']
df_Alpha1  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Alpha1']
df_Alpha2  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Alpha2']
df_Beta  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Beta']
df_Beta1  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Beta1']
df_Beta2  = df_analysis.loc[df_analysis['brain_oscillation'] == 'Beta2']
```

```python
df_Delta.to_csv('Data/df_Delta')
df_Theta.to_csv('Data/df_Theta')
df_Alpha.to_csv('Data/df_Alpha')
df_Alpha1.to_csv('Data/df_Alpha1')
df_Alpha2.to_csv('Data/df_Alpha2')
df_Beta.to_csv('Data/df_Beta')
df_Beta1.to_csv('Data/df_Beta1')
df_Beta2.to_csv('Data/df_Beta2')
```
